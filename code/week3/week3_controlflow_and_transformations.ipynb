## Conditionals

Now that we know how indentation works in Python, we can start defining logic conditions that determine how our program runs. In programming, conditions are governed by Boolean (or binary) logic which is the math of values that can be either true or false.  The name Boolean comes from the mathematician [Georges Boole](https://en.wikipedia.org/wiki/George_Boole) who developed the mathematical fundations of this method. In Python, Boolean values are written as `True` and `False` (note the capital letters, which differs from JavaScript and many other languages).


In Python we can test wether something is `True` or `False` by using the `if` statement. Say we want to execute a block of code only if a condition is true:

````python
condition = True
if condition:
    print("Condition is True")
print(condition)
````

Note the use of indentation to determine what happens when the condition is `True`. Again, differently from JavaScript we don&rsquo;t need a parenthesis after the `if` statement.

We can combine and operate on multiple Boolean values using "logical operators", which in Python have English names: `and`, `or` and `not`.

````python
condition1 = False
condition2 = True
if condition1 and condition2:
    print("The combined condition is True")
````

The `elif` and `else` statement can be placed after an `if` statement to test for different conditions, e.g.



````python
if condition1 and condition2:
    print("Both conditions true")
elif condition1 or condition2:
    print("At least one consition is true, maybe both")
else:
    print("Both are false")
````

We can test if the opposite of a condition by using the `not` operator, so



````python
if not condition1:
    print("Condition 1 is False")
else:
    print("Condition 1 is True")
````

````python
not False
````

The `not` operator placed in front of a Boolean expression, negates the expression so `True` becomes `False` and the other way around.



## (In)equality operators



`if` statements function based on a Boolean value, but we can use equality expressions to produce `True` or `False` results. For example consider the question:

-   *&ldquo;Is the value of `a` greater than 0?&rdquo;*

the answer to that is either yes or no, which in Python translates to either `True` or `False`.
We can write this as:



````python
a = -1
a < 0
````

We can use this in an `if` statement to drive the logic of the program:



````python
print("IS `a` GREATER THAN 0?")
if a > 0:
    print("The answer is YES (True)")
else:
    print("The answer is NO (False)")
````

Similarly to `>` (greater than) we can also use other operators to test between two elements:

-   `==` equality, `True` if the values on each side are equal, e.g. `a==0`.
-   `!=` inequality, `True` if the values on each side are different, e.g. `2 != 3` is `True`.
-   `<` less than, `True` if the value on the left is less than the value on the right, e.g. `4<0` is `False`.
-   `>=` greater or equal, `True` if the value on the left is greater or equal than the value on the right.
-   `<=` less or equal, `True` if the value on the left is less or equal to the value on the right.

Try to use them in the code cell below:

````python
value = -20
print(value)
if value >= 0 and value < 3:
    print('Value is between 0 and 2')
elif value >= 3:
    print('Value is greater or equal to 3')
else:
    print('Value is negative')
````

### Loops with conditionals

We can replace `if` with `while` (and omitting any `elif` or `else`) and this will result in a loop that will **repeat** a block of code **while** the condition is `True`. Repetition with this method is literally a "loop", because when the indented block is ended, execution loops back to beginning of the block and repeats it until a certain condition is verified. 

As an example, when using interactive scripts we will write `run()` at the end of our script (instead of `show()`), the py5canvas engine internally will do something similar to this

```Python
def run():
    setup()

    while window_is_open():
        draw()
        show_canvas_in_window()
```

Here the function `window_is_open()` returns `True` if the sketch window is open and `False` if you close the window. You define the `setup` and `draw` functions. Then the `setup` function is automatically called once and the `draw` function is called repeatedly in a loop until you close the window (note that the `window_is_open` and `show_canvas_in_window` functions do not exist and are just here for demonstration purposes).

Generally, `while` loops are not recommended in most situations since it is easy to create loops that never exit (e.g. `while True`) and this can lead to the your program hanging.

## Transformations



In the examples above we had to shift each shape to a position along the grid and make sure that the positions we use we are scaled based on the number of rows and columns we chose. We did so by passing two initial arguments to our functions (`x` and `y`) that specify the position on the grid. While this is practical with simple shapes and code, it can become cumbersome as complexity grows. Enter transformations, which allow us to move, rotate or scale drawing commands as we please.

As we mentioned in previous classes, we draw shapes on the canvas by using a coordinate grid with two axes, one vertical pointing down (the &ldquo;Y&rdquo; axis) and one horizontal pointing to the right (the &ldquo;X&rdquo; axis). In 2d, we specify positions with a pair of numbers called coordinates (often written as a pair of numbers in parentheses, as `(30, 100)`), which define a distance to move along each axis (horizontally and vertically) to reach a point from the position `(0, 0)`.



When you write `create_canvas(...)` or `background` in your code, you start drawing from the top left corner of your canvas. The drawing below shows the canvas in gray and a point relative to the origin, but adds a shift of a few pixels at the beginning of the drawing code by using the `translate` function so we can actually see the axes. 

This function is a transformation and we will see a few more next. We will use two user defined functions

-   `draw_axes` draws the coordinate axes and displays their length.
-   `draw_pos(x, y)` draws a point at the coordinates specified and displays its position.

The functions in the cell below are defined as an utility to demonstrate transformaions. The use the `text` command, which we have not covered yet, involving [formatted strings](https://www.w3schools.com/python/python_string_formatting.asp) and [keyword (named) arguments](https://www.w3schools.com/python/gloss_python_function_keyword_arguments.asp). Feel free to read the linked articles for more info on how this works, but we will cover these techniques in the future.

````python
def draw_axes():
    x, y = 0, 0
    stroke_weight(3)
    stroke(255, 0, 0)
    arrow((x, y), (x+width, y))
    stroke(0, 255, 0)
    arrow((x, y), (x, y+height))
    fill(0)
    no_stroke()
    circle((x, y), 7)
    fill(0)
    text_size(14)
    text_align(CENTER, CENTER)
    

def draw_pos(x, y):
    fill(0)
    circle(x, y, 3)
    text_size(14)
    text_align(LEFT, TOP)
    text(f"({x}, {y})", x+3, y+3)
````

````python
create_canvas(400, 400)
background(255)
# We use transformations here to see the arrows
translate(10, 10)
scale(0.9)


fill(0, 40, 90, 40)
rect(0, 0, width, height)
fill(0, 150)
text_size(25)
text('Canvas', [100, 100])

draw_axes()
draw_pos(0, 0)
draw_pos(width/2, height/2)
show()
````

One way of thinking of transformations is to think of operations that move (`translate`), rotate (`rotate`) or scale (`scale`) your current coordinate axes.



### Translations



A translation moves the position of points in space. Think about it as displacement, or viewing yourself from the top, walking around a room, while always facing in the same direction. Or think it adding the first number to all your X coordinates and the second number to all the coordinates.

For example if we write `translate(width/2, height/2)` we can read it as &ldquo;Move the coordinate axes by `width/2` horizontally and `height/2` vertically&rdquo;. Because we start at the top left corner, this will move the &ldquo;origin&rdquo; `(0, 0)` to the center of the canvas:

````python
create_canvas(400, 400)
background(200)
translate(width/2, height/2) # Comment this

fill(0, 150, 250)
circle(0, 0, 150)

draw_axes()
draw_pos(0, 0)
draw_pos(-100, -150)
show()
````

This could be useful if we want to draw things centered in the canvas, or centered at any position. The idea is do write your drawing code relative to the origin `(0, 0)` and then use `translate` before the code so the origin is shifted to the desired position. In the code above we could have written `circle(width/2, height/2, 150)` without needing to use `translate`, but drawing code relative to the circle would become more complicated and more difficult to think of thereafter.

Notice that after `translate` a point with negative coordinates is now visible, while without `translate` negative coordinates are outside the canvas and not visible.

### Scaling



Note that in the previous image, we don't see the ends of the coordinate axes anymore. This happens because we translated the origin of the canvas to the position specified with `translate` and now coordinates with values greater than `width/2` or `height/2` are outside the canvas. 

If we want to see the axes endpoints we can use `scale` to shrink everything a bit, e.g. `scale(0.9)` will make everything 90% of its original size:
Scaling enlarges or shrinks shapes and the coordinate axes by a specifed amount. An amount smaller than one (`<1`) shrinks, an amount greater than one (`>1`) enlarges, and `1` does nothing. 

Imagine multiplying all your coordinates by that number.

````python
create_canvas(400, 400)
background(200)
translate(width/2, height/2) # Comment this
scale(0.5)
fill(0, 150, 250)
circle(0, 0, 150)

draw_axes()
draw_pos(0, 0)

show()
````

We applied a scale less than one in the code, which shrinked the coordinate axes have. That means that our &ldquo;ruler&rdquo; has shrinked, so a distance of 100 pixels is seen on the canvas as 50 pixels.

The previous example shows a **uniform** scaling, it scales X and Y coordinates by the same amount. If we specify two numbers we can scale the horizontal and vertical coordinates by different amounts:

````python
create_canvas(400, 400)
rect_mode('center')
background(255)
translate(width/2, height/2) # Comment this
scale(0.4, 1)
fill(0, 150, 250)
square(0, 0, 200)
draw_axes()
draw_pos(0, 0)
show()
````

### Rotation



Rotation rotates points around the current origin. By default, rotation is specifed in radians, which measure the distance a point has traveled along a circle with radius 1



A half rotation is [$\pi = 3.14159265358979323...$](http://www.geom.uiuc.edu/~huberty/math5337/groupe/digits.html) (on and on forever...) radians, a full rotation is $2\pi$ radians. $2\pi$ us the number of times you need to wrap the diameter around a circle and it is [also known as $\tau$](https://www.mathnasium.com/blog/tau-constant-better-pi). In Py5canvas (and Processing/p5js) these values are available as constants `PI` and `TWO_PI`.

>NOTE this image shows a counterclockwise rotation, which is the mathematical convention. In Py5canvas the Y axis points down, so a positive rotation will be clockwise.

You can use the `radians` function to convert degrees to radians. e.g. `radians(90)` will give you the equivalent of 90 degrees in radians. 

````python
create_canvas(300, 300)
background(150)
#rect_mode(CENTER)
translate(width/2, height/2) # Comment this
rotate(radians(45)) # or PI/4
fill(0, 150, 250)
square(0, 0, 200)
draw_axes()
draw_pos(0, 0)
show()
print(radians(45))
````

You can also specify degrees directly by using `angle_mode(DEGREES)` at the beginning of your code.

Notice how our axes are now not oriented straight anymore. Now moving along the X means sliding along the rotated direction. 



Rotations happen about the origin and, by default, squares and rectangles are drawn from the top left, so the square is not centered. 

You can use `rect_mode(CENTER)` to tell Py5canvas that you want to draw rects from the center instead of the corner. You can then use `rect_mode(CORNER)` to revert). 



We can include `scale` to see the axes endpoints again, and use a non-uniform scale to squash thes square into a rectangle:

````python
create_canvas(300, 300)
background(150)
rect_mode(CENTER)
translate(width/2, height/2) # Comment this
rotate(radians(60)) # or PI/4
scale(0.6, 0.5)
fill(0, 150, 250)
square(0, 0, 200)
draw_axes()
draw_pos(0, 0)
show()
````

Notice that the rectangle is scaled along the coordinate axes **transformed** by the previous translation and rotation.

If we apply the non-uniform scale **before** the translation, we will get a more unexpected result:

````python
create_canvas(400, 400)
rect_mode('center')
background(255)
translate(width/2, height/2) # Comment this
scale(0.25, 1)
rotate(radians(45))
fill(0, 150, 250)
square(0, 0, 200)
draw_axes()
draw_pos(0, 0)
show()
````

This results in a transformation known as a &ldquo;shear&rdquo;, which also modifies angles in addition to coordinates. We will not cover this in detail during this module, but in practice, the rotation has a different effect because once we scaled non-uniformely circles are not circles become ellipses and rotations behave differently.



### Combining transformations - Order matters!



The example above shows how multiple transformations above can be concatenated and the result **depends on the order** in which this concatenation occurs.

Getting your head around the effects of combined transformations can be confusing and is often a problem for newcomers to computer graphics. One way to think about transformations is the "turtle" analogy, which is used in a didactic but powerfull programming language known as LOGO,
where you create drawings by telling a turtle seen from the top how to move.



The turtle faces along the positive x axis, we will load an image from disk and draw it with a function `draw_turle()` and move it with a function `forward(amount)` that translates it along the x axis


````python
turtle = load_image('./images/turtle.png')

def draw_turtle(size=70):
    rect_mode(CENTER)
    image(turtle, -size/2, -size/2, size, size)

def forward(amt):
    translate(amt, 0)
````

````python
create_canvas(400, 400)
background(205)

# Move origin to canvas center
translate(width/2, height/2)

# Initial turtle
draw_turtle()

# Make a move
rotate(radians(30))
forward(100)
draw_turtle()

rotate(radians(-110))
forward(150)
draw_turtle()

scale(0.5)
forward(150)
draw_turtle()

show()
````

You can keep track of points by using the `get_origin()` function and then drawing the series of points. It does not exist in Processing/Py5js and it returns the position of the origin in canvas coordinates. You will need the latest version of py5canvas to use it, which you can install (after activating the py5 environment!!!) with
```
pip install --upgrade py5canvas
```

````python
from importlib import reload
import py5canvas.canvas
reload(canvas)
from py5canvas import *

points = []

def draw_turtle(size=70):
    rect_mode(CENTER)
    image(turtle, -size/2, -size/2, size, size)
    points.append(get_origin())
    
create_canvas(400, 400)
background(205)

# Move origin to canvas center
translate(width/2, height/2)

# Initial turtle
draw_turtle()

# Make a move
rotate(radians(30))
forward(100)
draw_turtle()

# Another move
rotate(radians(-110))
forward(150)
draw_turtle()

# Also scale the turtle and move
scale(0.5)
forward(150)
draw_turtle()

# Reset transformation and draw
reset_matrix()
stroke(0, 128)
no_fill()
begin_contour()
for x, y in points:
    vertex(x, y)
end_contour()

show()
````

Here we see a few new concepts. 

First, we keep track of the positions by adding elements to a list with `.append(element)`. The append function is a "method" of the Python list. We have not seen this yet, as it is part of Object Oriented Programming, a topic we will cover slowly over the weeks. Think of the dot as indicating a property or functionality specific to the object (a list in this case). Appending to a list adds an element to the end of the list, so it increases its size by 1.

Also, note that we did not have to write `global points` inside the function to modify the global `points` variable. This is because we are changing the length of the list but not replacing the value that the `points` variable refers to.

Finally, we created a custom polygon by using a loop surrounded by `begin_contour()` and `end_contour()`. The loop iterates over `points` and specifies a vertex (a point) using the `vertex(x,y)` function. Because each point in `points` consists of two coordinates, the syntax `for x, y in points:` distributes the contents of each points into two varaibles `x` and `y`.



### Thinking of transformations, first person (local) view



The way in which we conceptualized tranformations above, can be compared to a &ldquo;top-down view&rdquo; game, where we are in a 2d world viewed from the top but we are controlling the character. The origin coincides with the character and the charcter faces towards the X axis



This method makes it easy to build articulated bodies, e.g. making something like a robot arm:


````python
def draw_link(length):
    stroke(0)
    no_fill()
    rect(length/2, 0, length, 60)
    draw_turtle() # Comment me out

create_canvas(400, 400)
rect_mode(CENTER)
background(230)

translate(30, height/2) # layout

# Draw arm
for i in range(5):
    draw_link(100)
    translate(100, 0)
    rotate(radians(40))
    scale(0.8)
show()
````

#### Thinking of transformations, god (global) view



Anther way to think of transformations is to think of manipulating objects on the canvas while the coordinate system and the origin remain fixed. Think of it as moving or rotating paper on a board when making a collage. The main drawback of this conceptualization is that you must think of the rder of transformations in &ldquo;reverse&rdquo;, starting from the objects/shapes we are drawing and going backwards.
This makes more sense if you think of transformations as **functions** that applied to what follows and return a transformed shape.  

E.g. earlier we used a function inside a function `print(radians(45))`, and we read this backwards: The function ~radians~ is applied to the value 45 and the function ~print~ prints the result. With a composition of transformations of this type:
```python
translate(...)
scale(...)
rotate(...)
draw_shape()
```
you can imagine this to be something equivalent to
```python
translate(scale(rotate(shape)))
```

where `rotate(shape)` gives a new rotated shape and `scale(rotate(shape))` scales the rotated shape, etc... So you read the transformations backwards, going from the shape, with each transformation being applied to the result of the previous.

One example where this conceptualization is useful is modeling a planetary system with the sun at the center and one (for now) planet revolving around it. 

We will think of it as follows:

We first move our origin (the turtle) to the center of the canvas (that is the top-down view described earlier) and draw the sun at the origin. Now let's take a god point of view and think of how we would place the planet with respect to the sun and origin:

-   We shift the planet from the origin.
-   We rotate the shifted planet around the origin by a certain amount.

To do this we write:

````python
create_canvas(400, 400)
background(0)
# Origin at center
translate(width/2, height/2)
# Sun
fill(255, 128, 0)
circle(0, 0, 40)
# orbit
no_fill()
stroke(255, 30)
circle(0, 0, 150)
no_stroke()
# Planet
rotate(radians(30)) #PI/2)
translate(150, 0)
fill(0, 150, 200)
circle(0, 0, 20)
show()
````

### Saving and resetting transformations



In the previous example, we have seen how to place a planet around the sun. But we want to add more planets with the same procedure we used earlier (shift and rotate), and all the drawing functions that follow the transformations will be alrady transformed.

E.g. if you add another planet to the previous canvas without somehow resetting your transformation (note that in a notebook we can keep on working on the same canvas)



````python
rotate(-PI/2)
translate(100, 0)
fill(0, 180, 100)
circle(0, 0, 10)
show()
````

This can be useful for example to add a "moon" to the planet. But what if we want to add multiple planets with potentially multiple moons per planet?

We can avoid these problems by using a construct known as the &ldquo;trasformation stack&rdquo;, which allows you to save (push) the current trasformation and restore (pop) the last saved transformation using the `push_matrix()` and `pop_matrix()` commands. The name &ldquo;matrix&rdquo; appears because transformations are usually encoded as a matrix (a grid) of numbers.

So to draw an arbitrarily large number of planets we can do:

````python
create_canvas(400, 400)
background(0)
# Origin at center
translate(width/2, height/2)
# Sun
fill(255, 128, 0)
circle(0, 0, 40)

for i in range(5):
    dist = 70 + i*30
    no_fill()
    stroke(255, 30)
    circle(0, 0, dist)
    no_stroke()
    push_matrix() # Save current transformation
    rotate(random(-PI, PI))
    translate(dist, 0)
    fill(0, 150, 200)
    circle(0, 0, random(5, 10))
    pop_matrix() # Restore last saved transformation
show()
#save('planets.svg')
````

An alternative and slightly more advanced **syntax** that you can use in Python and will not work in Javascript is the following:



````python
create_canvas(400, 400)
background(0)
# Origin at center
translate(width/2, height/2)
# Sun
fill(255, 128, 0)
circle(0, 0, 40)

for i in range(5):
    dist = 70 + i*30
    no_fill()
    stroke(255, 30)
    circle(0, 0, dist)
    no_stroke()
    with push_matrix(): # Save current transformation
        rotate(random(-PI, PI))
        translate(dist, 0)
        fill(0, 150, 200)
        circle(0, 0, random(5, 10))
show()
````

Here we use the indentation syntax, together with the keyword `width` to tell
the program that after all the indented code following `with push_matrix():`, we need to automatically restore the previous transformation with `pop_matrix()`.



### Back to grids



In the earlier examples, we have seen that we needed to carefully adjust coordinates to place our drawings in a grid.
With the tools we have learned we can now simplify the process, and just think of a drawing with respect to the origin and then transform it for each cell of a grid. Let&rsquo;s start by working on what we want to display in each cell and put it into a function.

We will work in a 300x300 canvas with center at the origin:



````python
tile_size = 300
create_canvas(tile_size, tile_size)
def draw_tile():
    rect_mode('center')
    no_stroke()
    fill(50)
    square(0, 0, 250)
    fill(255, 128)
    for i in range(30):
        circle(random(-100, 100),
               random(-100, 100),
               20)

translate(width/2, height/2)
draw_tile()
save('myimage.svg')
show()
````

Because the `draw_tile()` function uses randomness, it will be different each time we use it.
Now to the grid code, it is similar to the one we did previously, but we will will use `push_matrix` and `pop_matrix` to transform the same drawing function without needing to supply the coordinates:



````python
create_canvas(500, 500)

num_rows = 10
num_cols = 10
x_spacing = width / num_cols
y_spacing = height / num_rows

for i in range(num_rows):
    # Each row
    for j in range(num_cols):
        # Each column
        push_matrix()
        # Move to the tile top left corner
        translate(j*x_spacing, i*y_spacing)
        # Move to the center of the tile
        translate(x_spacing/2, y_spacing/2)
        # Scale to fit the grid
        scale(x_spacing/tile_size)
        # and draw
        draw_tile()
        pop_matrix()
show()
````

Because we put the transformation between `push_matrix()` and `pop_matrix()` each transformation works independently and does not &ldquo;add up&rdquo; to the previous. Note that we originally designed a random tile on a 300x300 (the 300 is stored in the `tile_size` variable) canvas with the origin shifted to the center. So our transformation code first moves the origin to the top left corner of the tile, then moves further to the center of the tile, and then scales the coordinates so that they fit in the (smaller) size of a tile. The amount of scaling is given by `x_spacing/tile_size`, where the scale will be smaller than 1 if `tile_size > x_spacing`.
