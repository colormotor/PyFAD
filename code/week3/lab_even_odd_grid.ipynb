__7) Checking for odd or even columns / rows__

We have a new arithmetic operator to learn that's very useful. Modulo `%` gives us the remainder after a division as a whole number. 

_Example 1:_ Imagine we divide 4 by 2, 4 is divisble by 2 so we won't have a remainder, the remainder is 0.   
_Example 2:_ 5 is not divisble by 2 however, 2 goes into 5 twice with 1 left over, so 5 modulo `%` 2  would give us 1.    

Using modulo `%` we can now start to check whether `i` or `j` are odd or even. Try and recreate the last three images above to understand how these following conditionals are working. Use one at a time to recreate the images above.

   If `j` is an even number (divisible by 2):

    ```Python

    ```

   If `i` is an even number (divisible by 2):

    ```python
    if (j%2 == 0 and i%2 == 0) or (j%2 != 0 and i%2 != 0):
    ```
