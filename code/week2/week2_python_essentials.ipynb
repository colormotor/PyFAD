Functions are extremely important in any programming language, **because they allow us to introduce modularity in the code**. #### Function syntax
The syntax we used to define the function is important to understand. It is called a **function declaration**.
We start with the keyword `def`, followed by the function&rsquo;s name and a set of comma-separated names inside parentheses. These names act like variables, but they are called &ldquo;arguments&rdquo; (or &ldquo;parameters&rdquo;). They allow the function to be flexible because the user can provide different values for the arguments each time the function is used.

After the declaration we have a colon (`:`) **followed by a newline and an indentation**. The indented block of code contains the definition of the function and what it will do.

    def my_function(argument1, argument2, ...):
        what the function will do

> This syntax is fundamental to Python and differs from languages such as JavaScript, where we define what a function does using curly braces (`{}`). Using indentation is useful to keep code tidy, since it graphically displays the relations between blocks of code, e.g. showing the code that is part of a function definition. We will see other examples of how indentation is used soon, but first some more on function definitions.

To use the function (often referred to as &ldquo;calling&rdquo; the function) we write its name followed by the arguments we supply to it, e.g.

    my_shape(10, 10, 50)

The arguments that are supplied to a function behave identically to variables, just they must be supplied by a user of the function. The name of the arguments is local to the function and is not necessarily important when calling the function, e.g. calling

    x = 10
    my_shape(x, 20, 30)

will work whatever name we gave to the arguments when defining the function.



### Scope and globals



The arguments to a function and any variable defined inside the function live inside the function (in the indented block) and are not available elsewhere. This behavior is normal and is called &ldquo;scope&rdquo;. The scope is like a house where a variable lives, and anybody outside cannot see the variable. For example this will give an error:



````python
def my_function_with_scope(arg):
    value2 = arg*2
    print(value2)

print(arg)
````

Because both `value` and `value2` are accessible only to the indented code block **inside** the function definition. 

Now say we define a variable **outside** of the function definition. The variable is then called a **global variable** or global for short. The scope of the variable contains also the function, so the code in the function is allowed to access its value. For example this will work:

````python
value = 10

def print_value():
    print(value)

print_value()
````

A function can also modify a global variable, but that cannot be done without some additional syntax, e.g. this will not work

````python
value1 = 5
value2 = 10

def increment():
    value1 = value1 + 1
    value2 = value2 + 1
    print(value1, value2)

increment()
````

To allow us to modify a global variable from within a function, we need to tell Python and it can be done with the `global` keyword followed by the variable names that we 
want to modify separated by commas. So the corrected code will be:

````python
value1 = 5
value2 = 10

def increment():
    global value1, value2
    value1 = value1 + 1
    value2 = value2 + 1
    print(value1, value2)

increment()
````

Generally, global variables should be used sparingly, unless you are working in an interactive sketch (which we will do next week).

## Sequences and iteration/loops

So far, we have done simple drawings by repeating commands. But this is probably much easier to do in a design software of choice. To make things more interesting, we need to introduce repetition/iteration/loops in the code, which in Python usually means thinking of collections or sequences of things. 

A Python sequence is an ordered collection of items (variables) that can be accessed by an index. The most common sequences in Python are "strings", "lists", "ranges", "tuples". 

- A "string" (type `str`) is a sequences of characters defined by surrounding text with quotes, e.g. `"hello"` or `'hello'`. 

- A "list" (type `list`) is a sequence of variables. Items in a list can be modified, removed or added. 
  A list can be defined as a comma separated sequence of variables enclosed by square brackets, 
  e.g. `["apple", "banana", "milk", "bread"]`.
 
- A "range" (type `range`) is a sequence of integers, it is similar to a list but it is faster when accessing one item after the other.
  E.g. `range(10)` gives us a sequence of ten numbers from `0` to `9` (note that we start counting from zero)

- A "tuple" (type `tuple`) is a sequence of variables similar to a list, but its elements cannot be modified.
  It can be defined as a comma separated sequence of variables, optionally enclosed by parentheses.
  E.g. `(0, 1)` or `x, y`.


Now, say that we want to draw 100 shapes instead of three as we did earlier. We could write 100 lines of code, but that would be tedious and error-prone and ai you think of the pattern we are **repeating** a drawing of a similar shape at different locations. Repeating an operation in programming is also known as a []"loop" or "iteration"](https://en.wikipedia.org/wiki/Iteration), and it is a fundamental concept of programming, which actually makes programming as useful as it is.

Python has a particular way of doing iteration, which is different from other programming you are using such as JavaScript. In Python, you typically iterate over sequences of items. Intuitively, imagine a repetitive task such as scanning the items in your basket at the till of a supermarket. You would take each item one by one and scan it. You can think of it as:

    For each item in the sequence, do something with the item.

In Python it could look something like:
```python
basket = ["apple", "banana", "milk", "bread"]
for item in basket:
    scan(item)
```
Where we would have to define the `scan` function to do something useful, e.g. print the item.

Note the definition of `basket` as a list of strings delimited by square brackets. When you see square brackets you know you are dealing with a list of things.

So you have the `for` keyword, followed by a variable name (in this case `item`) that will take the value of each item in the sequence (in this case `basket`) one by one. Similarly to function definitions, the indented block of code below the `for` statement defines what gets repeated for each item in the sequence, and the variable following the `for` is set to the value of the item for each step of the iteration. 

You can use the same procedure with any sequence-like object in Python, e.g we could count the number of characters in a string with a `for` loop:

````python
txt = "Hello world"
count = 0
for c in txt:
    print(count, c)
    count += 1 # This is a shorthand for i = i + 1
print(count, 'characters')
````

Note how we started from zero and added 1 at each iteration. 

The code above displayed an increasing number for each element in a sequence (a string in this case). We started counting from zero. The string has `11` characters and the last character has the number `10` starting from zero. A faster way in Python to query the length of a sequence is the built-in `len(sequence)` function. E.g. 

````python
len(txt)
````

The numbers, **starting from zero**, that we printed earlier next to each character, can be used to "index" single elements of the sequence using square brackets following the name of the variable. E.g.

````python
txt[0]
````

Gives us the first element in the sequence,`txt[1]` would give us the second and so forth. You of text as a sequence like



The sequence has a size and accessing an element with index greater or equal to the size will give an error:

````python
txt[11]
````

because the sequence has `len(txt)` elements but the index of the last element is `len(txt)-1` as we started to count from zero. This is a pattern that can be a bit confusing but is common to most programming languages you are likely to use during the course, e.g. Javascript and C++. Indexing sequences has many more features in Python but we will see this later in the module.

## Remap function



Earlier, we used calculations inside and outside of a loop to compute the coordinates of shapes. Py5canvas, provides a function that is useful for these situations called \`remap\` and it is equivalent to `map` in p5.js/Processing. You can watch a video by Dan Shiffman for a [great tutorial](https://www.youtube.com/watch?v=nicMAoW6u1g) on its use (just keep in mind that it is \`remap\` in Py5canvas, because Python uses \`map\` for other tasks).

\`remap\` is used to convert a value from one range to another range while maintaining proportional relationships. This is extremely useful for parametric designs and data visualization.

**Syntax:**

    remap(value, start1, stop1, start2, stop2)

-   `value`: The input value to remap
-   `start1`, `stop1`: The original range the value comes from
-   `start2`, `stop2`: The target range to map to

E.g.



````python
remap(0.5, 0.0, 1.0, 0, 200)
````

Play with the values to get a feel of this and think of it as something like this:



or as a unit conversion



Let&rsquo;s look at a re-adaptation of an example from earlier:



````python
create_canvas(400, 400)
background(0)
for i in range(n):
    fill(remap(i, 0, n-1, 0, 255)) # remap i to a number between 0 and 255
    rectangle(remap(i, 0, n, 0, width),
              0,
              20,
              random(0, height))
show()
````

Note that we use `n-1` for the input maximum in `remap`. The reason for this is that `range` will return numbers up to `n-1` and not up to `n`!

````python
# Example 1: Map vertical position to circle size
create_canvas(300, 300)
background(255)
n = 100
for i in range(n):
    # Map i to a y position between 0 and height
    y = remap(i, 0, n-1, 0, height)
    # Map y position to diameter in a given range
    w = remap(y, 0, height, 5, 200)
    ellipse(width/2, y, w*2, w)
show()
````

## Normalization 

A trick that can be handy is to remap values to a range between -1 and 1 or between 0 and 1. The reason this is handy is because when you multiply two values between -1 and 1 the result will always be between -1 and 1 (e.g. `0.2*0.5` is `0.1`). This becomes particularly useful when doing interpolation (which we are already doing with `remap`), a topic that we will cover in the following weeks. Let's look at an example:

````python
create_canvas(400, 400)
background(0)
n = 50
for i in range(n):
    t = remap(i, 0, n-1, -1, 1) # Map to -1, 1
    h = remap(t*t*t,
              -1, 1, 0, height) # Do something and then map back to the desired range
    x = remap(i, 0, n, 0, width) 
    rect(x, 0, 10, h)
show()
````

Here we did `t*t*t`, which is the same as "t to the power of 3" ($t^3$), which in Python can be written more concisely as `t**3`. We can use this to create non-linear color gradients in combination with the HSB (Hue Saturation Brightness) color model as opposed to RGB (see the [Tutorial on colors](./../week1/noc_color.ipynb) from last week).

````python
create_canvas(400, 400)
background(0)
no_stroke()
color_mode(HSB)
n = 30
for i in range(n):
    t = remap(i, 0, n-1, 0, 1) # Map to 0, 1
    v = remap(t**3, # Try values < 1 for the power
              0, 1, 0, 255) # Do something and then map back to the desired range
    w = remap(i, 0, n, width, 10)
    fill(v, 255, 255)
    circle(width/2, height/2, w)
show()
````

# Nested `for` loops

So fare we have used single `for` loops to create repetition. Note how we put the code that gets repeated in an **indented block following a `:`**. This code can contain more loops, and nesting one more for loop inside another can be used to create a grid.

Here is an example of this 

````python
create_canvas(400, 400)
background(0)

stroke_weight(2)

num_rows = 5
num_cols = 5

# Width and height of each cell
w = width/num_cols
h = height/num_rows

for i in range(num_rows):
    # For each row
    for j in range(num_cols):
        # For each column
        # print('i:',i, 'j:',j) # Uncomment to print row and column
        fill(random(0, 255))
        x = remap(j, 0, num_cols, 0, width) 
        y = remap(i, 0, num_rows, 0, height)
        rect(x, y, w, h)
show()
````

Note two things:

- This time we set input maximum for the ranges to `num_cols` and `num_rows` and not `num_cols-1` and `num_rows-1`. This leaves the space for the extra row and columns that would disappear if we did the former. 
- This kind of looks like a zoomed in version of an image. Indeed we can think of the pixels in an image as a grid of this type, and we will do similar loops to directly manipulate pixels in the future.
