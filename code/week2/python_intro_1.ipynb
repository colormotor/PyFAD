## Variables



A variable is like a storage box where you can keep information that you want to use later in your program. You give the box a name (the variable name), and then you can put something inside it, e.g. a number or some text (a string). When you want to use that information again, you just refer to the box by its name, and Python will give you whatever is inside.

E.g. let&rsquo;s define two variables:



```python
a = 10
b = "Some text"
```

In notebooks, the last line of a code cell is shown under the code shell so if we write



```python
a
```

we get the result. We can also print the contents of one or more variables (separated by commas if more than one) at any point in the cell by using the `print` statement:



```python
print(a, b)
```

The type of information contained in a variable is called it&rsquo;s `type`. You can query it by using the `type` command, e.g.:



```python
type(a)
```

Python infers what type of information a variable contains automatically when it is created. This is called &ldquo;dynamic typing.&rdquo;. You can change the type by re-assigning anther value to the same variable, e.g.:



```python
a = 10.3
type(a)
```

Here is a list of variable types we will encounter today:

-   `string`, strings of text. It is delimited by double or single quotes (e.g. `"hello"` or `'Hello world'`).
-   `int`, integers (e.g. `0`, `100`, `-420`).
-   `float`, floating point numbers (with a fractional part, e.g. `0.3`, `-1.22321112`).
-   `tuple`, a sequence of variables. It is delimited by parentheses (e.g. `(0, 1, 2)`, `("a", "b")`) or simply a series of values separated by commas (e.g. `0, 1, 2`).
    
    Some types can be converted between each other by using the type name as a function, e.g. `str(10)` will convert 10 to a string, or `int("10")` will convert the string to an int (if possible).



### Example



As an example of why variables can be useful, say we want to draw three or more circles but we want to experiment with the radius and positioning of the circles. We could &ldquo;hard code&rdquo; the positions and radii for each circle:



```python
create_canvas(300, 300)
circle(100, 100, 30)
circle(100, 150, 20)
circle(100, 200, 10)
show()
```

However, we realize that we want the circles to be centered, and now we need to change the &ldquo;x&rdquo; coordinate of all circles. A better way to do this is to store tha positions, and why not the radius, in variables, e.g.:



```python
create_canvas(300, 300)
x = 150
y = 100
r = 30
circle(x, y, r)
circle(x, y+r*2, r*0.5)
circle(x, y+r*3, r*0.25)
show()
```

Here you can see that we also can apply arithmetic operations on the variables, where `+, -, *, /` stand for addition, subtraction, multiplication and division.



#### Properties in Py5Canvas



Py5Canvas provides a number of special variables called &ldquo;properties&rdquo;, which give us access to information that might be useful for drawing. E.g. what if we wanted to change the size of the canvas in the example above and automatically have the circles centered horizontally? We can access the `width` property, which gives us the width of the current canvas:



```python
create_canvas(500, 300)
x = width/2
y = 100
r = 30
circle(x, y, r)
circle(x, y+r*2, r*0.5)
circle(x, y+r*3, r*0.25)
show()
```

One thing to watch out for is that we can overwrite the `width` variable and then the value won&rsquo;t be updated until we create a new canvas with `create_canvas` or `size`.

## Functions, modularity and scope



We have seen that Py5Canvas gives us access to a number of &ldquo;functions&rdquo;, commands that allow to draw pictures on a screen and are distinguished by a name, followed by an opening and closing parentheses containing a number of parameters called **arguments**, with a syntax of the type `name_of_function()` or `name_of_function(argument1, argument2)`.

Functions are extremely important in any programming language, because they allow us to introduce modularity in the code. Like variables, functions have a name, but the name corresponds to an action/command instead of only containing information.

As a practical example, we may want to experiment with using shapes different from circles, so we could &ldquo;define&rdquo; our own function to do so:



```python
def my_shape(x, y, radius):
    circle(x, y, radius)
    rect_mode('center')
    rect(x, y, radius, radius)
```

And then use the function similarly to what we did with the circles:



```python
create_canvas(500, 300)
x = width/2
y = 100
r = 30
my_shape(x, y, r)
my_shape(x, y+r*2, r*0.5)
my_shape(x, y+r*3, r*0.25)
show()
```

The syntax we used to define the function is very important to understand. It is called a **function declaration**.
We start with the keyword `def`, followed by the function&rsquo;s name and a set of comma-separated names inside parentheses. These names act like variables, but they are called &ldquo;arguments&rdquo; (or &ldquo;parameters&rdquo;). They allow the function to be flexible because the user can provide different values for the arguments each time the function is used.

After the declaration we have a colon (`:`) followed by a newline and an indentation. The indented block of code contains the definition of the function and what it will do.

    def my_function(argument1, argument2, ...):
        what the function will do

This syntax is fundamental to Python and differs from languages such as JavaScript, where we define what a function does using curly braces (`{}`). Using indentation is useful to keep code tidy, since it graphically displays the relations between blocks of code, e.g. showing the code that is part of a function definition. We will see other examples of how indentation is used soon, but first some more on function definitions.

To use the function (often referred to as &ldquo;calling&rdquo; the function) we write its name followed by the arguments we supply to it, e.g.

    my_shape(10, 10, 50)

The arguments that are supplied to a function behave identically to variables, just they must be supplied by a user of the function. The name of the arguments is local to the function and is not necessarily important when calling the function, e.g. calling

    x = 10
    my_shape(x, 20, 30)

will work whatever name we gave to the arguments when defining the function.



### Scope and globals



The arguments to a function and any variable defined inside the function live inside the function (in the indented block) and are not available elsewhere. This behavior is normal and is called &ldquo;scope&rdquo;. The scope is like a house where a variable lives, and anybody outside cannot see the variable. For example this will give an error:



```python
def my_function_with_scope(value):
    value2 = value*2
    print(value2)

print(value2)
```

Because both `value` and `value2` are accessible only to the indented code block in the function definition. 

Now say we define a variable **outside** of the function definition. The variable is then called a **global variable** or global for short. The scope of the variable contains also the function, so the code in the function is allowed to access its value. For example this will work:

```python
value = 10

def print_value():
    print(value)

print_value()
```

A function can also modify a global variable, but that cannot be done without some additional syntax, e.g. this will not work

```python
value1 = 5
value2 = 10

def increment():
    value1 = value1 + 1
    value2 = value2 + 1
    print(value1, value2)

increment()
```

To allow us to modify a global variable from within a function, we need to tell Python and it can be done with the `global` keyword followed by the variable names that we 
want to modify separated by commas. So the corrected code will be:

```python
value1 = 5
value2 = 10

def increment():
    global value1, value2
    value1 = value1 + 1
    value2 = value2 + 1
    print(value1, value2)

increment()
```

Generally, global variables should not be used unless you are working in an interactive sketch (which we will see soon).

## Conditionals



Now that we know how indentation works in Python, we can start defining logic conditions that determine how our program runs. In programming, conditions are governed by Boolean (or binary) logic which is the math of values that can be either true or false.  The name Boolean comes from the mathematician [Georges Boole](https://en.wikipedia.org/wiki/George_Boole) who developed the mathematical fundations of this method. In Python, Boolean values are written as `True` and `False` (note the capital letters, which differs from JavaScript and many other languages).

In Python we can test wether something is `True` or `False` by using the `if` statement. Say we want to execute a block of code only if a condition is true:



```python
condition = True
if condition:
    print("Condition is True")
print(condition)
```

Note the use of indentation to determine what happens when the condition is `True`. Also, differently from JavaScript we don&rsquo;t need a parenthesis after the `if` statement.

We can combine and operate on multiple Boolean values using &ldquo;logical operators&rdquo;, which in Python have English names: `and`, `or` and `not`.



```python
condition1 = True
condition2 = True
print("Conditions", condition1, condition2)
if condition1 or condition2:
    print("The combined condition is True")
```

The `elif` and `else` statement can be placed after an `if` statement to test for different conditions, e.g.



```python
if condition1 and condition2:
    print("Both true")
elif condition1 or condition2:
    print("One true"):
else:
    print("None true")
```

We can test if the opposite of a condition by using the `not` operator, so



```python
if not condition1:
    print("Condition 1 is False")
else:
    print("Condition 1 is True")
```

The `not` operator placed in front of a Boolean expression, negates the expression so `True` becomes `False` and the other way around.



## (In)equality operators



`if` statements function based on a Boolean value, but we can use equality expressions to produce `True` or `False` results. For example consider the question:

-   *&ldquo;Is the value of `a` greater than 0?&rdquo;*

the answer to that is either yes or no, which in Python translates to either `True` or `False`.
We can write this as:



```python
a > 0
```

and use it in an `if` statement to drive the logic of the program:



```python
if a > 0:
    print("The answer is YES")
else:
    print("The answer is NO")
```

Similarly to `>` (greater than) we can also use other operators to test between two elements:

-   `==` equality, `True` if the values on each side are equal, e.g. `a==0`.
-   `!=` inequality, `True` if the values on each side are different, e.g. `2 != 3` is `True`.
-   `<` less than, `True` if the value on the left is less than the value on the right, e.g. `4<0` is `False`.
-   `>=` greater or equal, `True` if the value on the left is greter or equal than the value on the right.
-   `<=` less or equal, `True` if the value on the left is less or equal to the value on the right.
